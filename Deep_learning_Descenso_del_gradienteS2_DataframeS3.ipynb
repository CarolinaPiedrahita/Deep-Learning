## **¿Qué es el algoritmo de descenso de gradiente?**

El algoritmo de aprendizaje automático más utilizado es probablemente el descenso de gradiente. El propósito fundamental del algoritmo es reducir los errores en la medida de lo posible. El propósito del algoritmo de descenso de gradiente es reducir la función de costo de un modelo. Los significados de las palabras 'Gradiente' y 'Descenso' nos permiten afirmar esto (Galea & Capelo, 2018).

## **¿Cómo funciona el descenso de gradiente?**


Primeramente, es crucial recordar que el algoritmo de descenso de gradiente es un proceso iterativo, al igual que la mayoría de los procesos de aprendizaje automático. Suponiendo que tiene la función de costo para un modelo de regresión lineal simple como j(w,b) donde j es una función de w y b, el algoritmo de descenso de gradiente funciona de tal manera que comienza con una conjetura aleatoria inicial para w y b. El algoritmo continuará ajustando los parámetros w y b con el fin de maximizar la función de costo, j. La selección de los valores iniciales en la regresión lineal no tiene gran importancia. Cero es una opción común (Jang, Wu, & Jha, 2017).

# **¿Qué es la tasa de aprendizaje?**

El factor que determina la magnitud de los pasos que se toman en la dirección del mínimo local es la tasa de aprendizaje. Determina cuán rápido el algoritmo avanza hacia los valores ideales de la función de costo. La elección de la tasa de aprendizaje (γ), por lo tanto, es crucial y afecta significativamente la eficacia del algoritmo.  (Jang, Wu, & Jha, 2017).

```python
# Creamos la función a optimizar y la grafica.
funcion = lambda x: x**2 + 1

# Se crean 200 valores espaciados entre x=-10 y x=10
x = np.linspace(-10, 10, 200)

# Grafica la función.
plt.plot(x, funcion(x))
plt.title("Función a Optimizar")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

# Creamos la función de la derivada de x.
def derivada(x):
    return 2 * x  # Regresa la derivada de la función inicial

# Creamos la función del gradiente
def gradiente(x_inicio, precision, l_r):
    # Se crean dos listas para los valores de x,y
    # que se utilizarán después para la visualización puntual.
    x_list, y_list = [x_inicio], [funcion(x_inicio)]

    # Utilizamos el bucle While para hacer el recorrido
    while True:
        # Se obtiene el valor de la pendiente de la derivada
        # debido a que necesitamos un descenso negativo (hacia el mínimo)
        # usamos el signo menos '-'
        d_x = -derivada(x_inicio)

        # Obtenemos el nuevo valor de x sumando el valor previo de x más la multiplicación del l_r por la derivada.
        x_inicio = x_inicio + (l_r * d_x)

        # Añadimos el nuevo valor de x a la lista para posteriormente visualizarlo.
        x_list.append(x_inicio)

        # Añadimos el nuevo valor de y a la lista para posteriormente visualizarlo.
        y_list.append(funcion(x_inicio))

        # Salimos del bucle cuando se cumpla la condición.
        if abs(x_list[-1] - x_list[-2]) <= precision:
            break

    # Visualizamos los resultados.
    print("El mínimo local se produce en: " + str(x_inicio))
    print("Número de pasos: " + str(len(x_list)))

    # Visualizamos la gráfica del descenso del gradiente
    plt.figure(figsize=(14, 6))

    # Gráfico de la función inicial
    plt.subplot(1, 2, 1)
    plt.plot(x, funcion(x), label='Función')
    plt.scatter(x_list, y_list, color='red')
    plt.plot(x_list, y_list, label='Descenso del Gradiente', color='blue')
    plt.title("Función y Descenso del Gradiente")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)

    # Gráfico del descenso del gradiente paso a paso
    plt.subplot(1, 2, 2)
    plt.scatter(x_list, y_list, color='blue')
    plt.plot(x_list, y_list, color='blue')
    plt.title("Descenso del Gradiente Paso a Paso")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

###Solución

Importamos librerias

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Explicación del Código | Definición de la función y su derivada:**

f(x) es la función objetivo.

df(x) es la derivada de la función objetivo.

**Implementación del descenso del gradiente:**

gradient_descent es una función que realiza el algoritmo de descenso del gradiente.

Comienza en un punto inicial starting_point.

En cada iteración, calcula el gradiente, actualiza el valor de x y almacena los valores de x en x_history.

El proceso se detiene cuando la diferencia entre x nuevo y el anterior es menor que la precision o se alcanza el número máximo de iteraciones max_iterations.

**Parámetros:**
starting_points contiene los puntos iniciales.
learning_rates contiene las tasas de aprendizaje.
precision es la precisión deseada.
max_iterations es el número máximo de iteraciones permitidas.

**Graficar la función y los puntos obtenidos:**

Se grafica la función f(x) = $\large\  { (x)^{2}+5x+6} $, se hace la prueba con varios valores de $\large\  { (x)} $.

Para cada combinación de punto inicial y tasa de aprendizaje, se realiza el descenso del gradiente y se grafican los puntos obtenidos.

Se muestra el valor de x donde se alcanza el mínimo, el valor correspondiente de f(x) y el número de pasos necesarios.

```python


# Definir la función
def f(x):
    return (x)**2 + 5*x + 6

# Definir la derivada de la función
def df(x):
    return 2*x + 5

# Implementar el descenso del gradiente
def gradient_descent(starting_point, learning_rate, precision, max_iterations):
    x = starting_point
    x_history = [x]
    for i in range(max_iterations):
        gradient = df(x)
        new_x = x - learning_rate * gradient
        x_history.append(new_x)
        if abs(new_x - x) < precision:
            break
        x = new_x
    return new_x, f(new_x), x_history, i

# Parámetros
starting_points = [-10, 0, 10]
learning_rates = [0.01, 0.1, 0.2]
precision = 0.0001
max_iterations = 1000

# Graficar la función
x = np.linspace(-10, 10, 400)
y = f(x)
plt.plot(x, y, label='f(x)')

colors = ['r', 'g', 'b']

for i, starting_point in enumerate(starting_points):
    for j, learning_rate in enumerate(learning_rates):
        x_min, y_min, x_history, steps = gradient_descent(starting_point, learning_rate, precision, max_iterations)
        plt.scatter(x_history, [f(x) for x in x_history], color=colors[i], label=f'x0={starting_point}, lr={learning_rate}')
        print(f'Inicio en x0={starting_point}, lr={learning_rate}: x_min={x_min}, y_min={y_min}, pasos={steps}')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Descenso del Gradiente')
plt.show()
```

**En la gráfica se observa que:**



**Puntos de Inicio (x0) y Learning Rates (lr):**

Se probaron diferentes puntos de inicio (𝑥0=−10,0,10; x0=−10,0,10) y diferentes tasas de aprendizaje (𝑙𝑟=0.01,0.1,0.2 ;lr=0.01,0.1,0.2).

**Convergencia y Color de los Puntos:**

Los puntos en la gráfica muestran cómo se comportó el descenso del gradiente desde distintos puntos de inicio y con diferentes tasas de aprendizaje. Cada conjunto de puntos tiene un color distinto, representando una combinación específica de 𝑥0  y 𝑙𝑟

**Gráfico de la Función:**

La curva azul representa la función $\large\  { (x)^{2}+5x+6} $, haga la prueba con varios valores de $\large\  { (x)} $ . Observamos que la función es una parábola que abre hacia arriba, lo que implica que tiene un mínimo global.

**Comportamiento del Descenso del Gradiente:**

*Los puntos rojos* (inicio en 𝑥0=−10 x0=−10) con diferentes tasas de aprendizaje convergen hacia el mínimo. Notamos que las tasas de aprendizaje más altas (0.1 y 0.2) convergen más rápido.

*Los puntos verdes* (inicio en 𝑥0=0) también convergen hacia el mínimo, nuevamente con tasas de aprendizaje más altas convergiendo más rápido.

*Los puntos azules* (inicio en 𝑥0=10;x0=10) muestran un comportamiento similar, con tasas de aprendizaje más altas acelerando la convergencia.

**Mínimo de la Función:**

Al observar la gráfica, todos los caminos de descenso del gradiente tienden a converger hacia el mismo punto mínimo de la función. Esto se espera, ya que la función f(x) tiene un único mínimo global (Galea & Capelo, 2018).

**Número de Pasos para Alcanzar el Mínimo:**

Podemos inferir que una tasa de aprendizaje más alta reduce el número de pasos necesarios para alcanzar el mínimo, pero también puede hacer que el algoritmo oscile si es demasiado alta (Galea & Capelo, 2018).

**Una tasa de aprendizaje más alta** (0.2) converge más rápido, alcanzando el mínimo en menos pasos (19 a 22 pasos) en comparación con tasas más bajas (0.01), que requieren muchos más pasos (308 a 388 pasos).

**Los valores iniciales también influyen**, pero no tanto como la tasa de aprendizaje. Por ejemplo, con lr=0.1, el algoritmo converge entre 39 y 46 pasos, independientemente de si se empieza en -10, 0 o 10.

**En resumen, podemos concluir que:**

Diferentes combinaciones de puntos de inicio y tasas de aprendizaje efectivamente alcanzan el mínimo de la función  $\large\  { (x)^{2}+5x+6} $.

Tasas de aprendizaje más altas aceleran la convergencia, pero deben ser elegidas con cuidado para evitar la oscilación (Galea & Capelo, 2018).

El descenso del gradiente es robusto y converge al mismo mínimo global de la función  independientemente del punto de inicio, aunque la rapidez de convergencia depende de la tasa de aprendizaje elegida (Galea & Capelo, 2018).
